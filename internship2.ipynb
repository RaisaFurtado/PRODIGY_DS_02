```python
import pandas as pd

# Load the dataset
data = pd.read_csv('surat_uncleaned.csv')

# Preview the first few rows of the dataset
print(data.head())
```

```python
# Get basic information about the dataset
print(data.info())

# Check for missing values
print(data.isnull().sum())

# Get descriptive statistics
print(data.describe())
```

```python
# Drop rows with missing values (if any)
data_cleaned = data.dropna()

# Optionally, you could fill missing values with a specific value
# data_cleaned = data.fillna(value={'ColumnName': 'DefaultValue'})

# Preview cleaned data
print(data_cleaned.head())
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='property_name', data=data_cleaned)
plt.title('Count Plot of property_name')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='areaWithType', data=data_cleaned)
plt.title('Count Plot of areaWithType')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='square_feet', data=data_cleaned)
plt.title('Count Plot of square_feet')
plt.show()
```

```python

sns.histplot(data_cleaned['property_name'], kde=True)
plt.title('Distribution of property_name')
plt.show()
```

```python

sns.histplot(data_cleaned['areaWithType'], kde=True)
plt.title('Distribution of areaWithType')
plt.show()
```

```python

sns.histplot(data_cleaned['square_feet'], kde=True)
plt.title('Distribution of square_feet')
plt.show()
```

```python
# Save the cleaned data to a new CSV file
cleaned_data_path = 'surat_cleaned.csv'
data_cleaned.to_csv(cleaned_data_path, index=False)

print(f"Cleaned data saved to {cleaned_data_path}")
```
